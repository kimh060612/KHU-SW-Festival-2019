```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
DATA = pd.read_csv("./New_train_V2.csv")
```

```python
train = DATA.copy()
```

```python
DBNO_heals_bins = [-1,10,20,30,40,50,80]
DBNO_heals_label = ["0-10 heals","11-20 heals","21-30 heals","31-40 heals","41-50 heals","50+ heals"]

DBNO_revive_bins = [-1,3,6,12,20]
DBNO_revive_label = ["0-3 revive","4-6 revive","7-12 revive","13-20 revive", "20+ revive"]

WD_numG_bins = [-1,20,40,80,120]
WD_numG_labels = ["0-20 NG","21-40 NG","41-80 NG","81-100 NG"]

WD_revive_bins = [-1,2,6,12,25,40]
WD_revive_labels = ["0-2 RE","3-6 RE","7-12 RE","13-25 RE","26-40 RE"]

WD_RD_bins = [-1,1000,5000,12000,20000,30000,50000]
WD_RD_label = ["0-1K RD","1K-5K RD","5K-12K RD","12K-20K RD","20K-30K RD","30K+ RD"]

WD_RK_bins = [-1,1,3,5,18]
WD_RK_labels = ["0-1 RK","2-3 RK","4-5RK","6-18RK"]

WD_SD_bins = [-1,100,300,1000,2500,3600]
WD_SD_labels = ["0-100 SD","100-300 SD","300-1K SD","1K-2.5K SD","2.5K+ SD"]

WD_TK_bins = [-1,1,3,5]
WD_TK_labels = ["0-1 TK","2-3 TK","4-6 TK"]

WD_VehicleD_bins = [-1,1,3,5]
WD_VehicleD_labels = ["0-1 VD","2-3 VD","4-5 VD"]

WD_WeaponA_bins = [-1,10,30,60,300]
WD_WeaponA_labels = ["0-10 WA","11-30 WA","31-60 WA","61+ WA"]
```
